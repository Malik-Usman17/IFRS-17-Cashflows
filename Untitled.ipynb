```python
def sum_at_risk_value(death_benefit, sum_assured, CV_BoY_after_premium_val, policy_month_val, total_policy_month):
    if (death_benefit == 2 and policy_month_val < total_policy_month):
        val = max(0, sum_assured - CV_BoY_after_premium_val)
        return val
    else:
        return 0
```

```python
def mortality_charge_value(CV_BoY_after_premium_val, sum_at_risk_val, mortality_rate_val, mortality_charge_loading):
    if (CV_BoY_after_premium_val <= 0):
        return 0
    else:
        value = mortality_rate_val * sum_at_risk_val * (mortality_charge_loading/100)
        return round(value)
    
mortality_charge_value(238679, 5946, 0.000938, 125)
```

```python
def admin_charge_value(CV_BoY_after_premium_val, admin_charges, admin_inflation, policy_year_val):
    if(CV_BoY_after_premium_val <= 0):
        return 0
    else:
        val = admin_charges * (1 + (admin_inflation/100)) ** (policy_year_val - 1) / 12
        return val
```

```python
def cash_value_EoY(CV_BoY_after_premium_val, mortality_charge_value, admin_charge_val):
    try:
        value = round(CV_BoY_after_premium_val - mortality_charge_value - admin_charge_val)
    except Exception as e:
        value = 0
    return value
```

```python
def investment_income(growth_rate_val, cash_value_eoy_val):
    value = round(cash_value_eoy_val * (growth_rate_val/100))
    return value
```

```python
def cash_value_after_investment_income(cash_value_eoy, investment_income):
    value = cash_value_eoy + investment_income
    return value
```

```python
def imc(cvai_income_val, imc_val):
    value = (cvai_income_val * (imc_val/100))/12
    return round(value)
    
#imc(237854, 1.75)
```

```python
def cash_val_eoy_net_of_imc(policy_month_val, total_policy_val, cvai_income_val, imc_val):
    if (policy_month_val > total_policy_val):
        return 0
    else:
        value = cvai_income_val - imc_val
        return value
```

```python
def sum_at_risk_eoy(policy_month_val, total_policy_month, death_benefit_val, cv_eoy_net_of_imc, sum_assured_val):
    if (policy_month_val >= total_policy_month):
        value = 0
    else:
        if(death_benefit_val == 2):
            value = max(sum_assured_val - cv_eoy_net_of_imc, 0)
        else:
            value = 0
    return value
```

```python
def calculate_policy_month(commencement_date, total_policy_months, start_date):
		policy_month_col = [months_into_days(commencement_date, start_date)]
		total_iterations = total_policy_months - policy_month_col[0]
		for i in range(1, total_iterations):
			if ((policy_month_col[i-1] + 1) < total_policy_months):
				policy_month_col.append(policy_month_col[i-1] + 1)
		return policy_month_col
```

```python
def cal_policy_year_value(month):
    try:
        result = int(month / 12) + 1
    except ZeroDivisionError:
        result = 0
    return result
```

```python
def calculate_policy_year(policy_month):
		policy_year_col = []
		for i in range(len(policy_month)):
			policy_year_col.append(cal_policy_year_value(policy_month[i]))
		return policy_year_col
```

```python
def calculate_age(age, policy_month, policy_year):
    age_column = []
    for i in range(len(policy_month)):
        if policy_month[i] > 0:
            age_val = (age + policy_year[i]) - 1
            age_column.append(age_val)
        else:
            age_column.append(0)
    return age_column
```

```python
from datetime import datetime, timedelta

def months_into_days(comm_date, start_date):
    date_a = datetime.strptime(comm_date, "%m/%d/%Y")
    date_b = datetime.strptime(start_date, "%m/%d/%Y")
    diff_in_days = (date_b - date_a).days
    result = round(diff_in_days / 30) + 1
    return result

months_into_days("4/28/2015", "1/1/2021")
```

```python
from datetime import datetime, timedelta

def age_at_comm(comm_date, insure_dob):
    date_a = datetime.strptime(comm_date, "%m/%d/%Y")
    date_b = datetime.strptime(insure_dob, "%m/%d/%Y")
    val_in_days = (date_a - date_b).days + 1
    value_in_years = round((val_in_days /365.25))
    return value_in_years

age_at_comm("4/28/2015", "5/25/1965")
```

```python
from datetime import datetime

def difference_in_years(date1, date2):
    try:
        # Convert the input strings to datetime objects
        date1_obj = datetime.strptime(date1, "%m/%d/%Y")
        date2_obj = datetime.strptime(date2, "%m/%d/%Y")
        
        # Calculate the difference in years
        time_delta = date2_obj - date1_obj
        difference_in_years = time_delta.days / 365.25  # accounting for leap years
        
        return round(difference_in_years)
    except ValueError:
        return "Invalid date format. Please use the format 'YYYY-MM-DD'."

# Example usage:
# date1 = input("Enter the first date (YYYY-MM-DD): ")
# date2 = input("Enter the second date (YYYY-MM-DD): ")
result = difference_in_years("5/25/1965", "4/28/2015")
print("Difference in years:", result)
```
