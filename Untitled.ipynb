```python
def policy_month(commencement_date, total_policy_months, start_date):
    policy_month_col = [months_into_days(commencement_date, start_date)]
    #policy_month_col[0] = months_into_days(commencement_date, start_date)
    #start_month_count = months_into_days(commencement_date, total_policy_months)
    total_iterations = total_policy_months - policy_month_col[0]
    for i in range(1, total_iterations):
        #print(i)
        if (policy_month_col[i-1] + 1) < total_policy_months:
            policy_month_col.append(policy_month_col[i-1] + 1)
    return policy_month_col
```

```python
def base_premium_col(policy_month, total_pol_month, frequency, prem_frequency, basic_premium):
    iteration = total_pol_month - policy_month[0]
    base_prem_list = [base_premium_first_val(policy_month[0], total_pol_month, prem_frequency, basic_premium, frequency)]
    for i in range(1, iteration):
        try:
            if (policy_month[i] - 1) % frequency == 0:
                val = 1
            else:
                val = 0
            val = val * basic_premium
            base_prem_list.append(val)
        except Exception as e:
            val = 0
            base_prem_list.append(val)
    return base_prem_list
```

## Calculating Allocated Percentage Column

```python
import pandas as pd
def calculating_allocated_percentage(policy_year):
    allocation_list = pd.read_csv("data/assumptions/301/allocation.csv", header=None).squeeze().tolist()
    allocation_percentage_list = []
    for i in range(len(policy_year)):
        year_val = policy_year[i]
        value = allocation_list[year_val - 1] * 100
        allocation_percentage_list.append(value)
    return allocation_percentage_list
```

### Calculating Allocated Premium Column

```python
def calculating_allocated_premium(base_premium, allocated_percentage):
    allocated_premium_list = []
    for i in range(len(base_premium)):
        value = (base_premium[i] * allocated_percentage[i]) / 100   #allocated_percentage ki value % mai hai jbhe / by 100
        allocated_premium_list.append(value)
    return allocated_premium_list
```

```python
def calculating_unallocated_premium(base_premium, allocated_permium):
    unallocated_premium_list = []
    for i in range(len(base_premium)):
        value = base_premium[i] - allocated_permium[i]
        unallocated_premium_list.append(value)
    return unallocated_premium_list
```

```python
def sum_at_risk_value(death_benefit, sum_assured, CV_BoY_after_premium_val, policy_month_val, total_policy_month):
    if (death_benefit == 2 and policy_month_val < total_policy_month):
        val = max(0, sum_assured - CV_BoY_after_premium_val)
        return val
    else:
        return 0
```

```python
def mortality_charge_value(CV_BoY_after_premium_val, sum_at_risk_val, mortality_rate_val, mortality_charge_loading):
    if (CV_BoY_after_premium_val <= 0):
        return 0
    else:
        value = mortality_rate_val * sum_at_risk_val * (mortality_charge_loading/100)
        return round(value)
    
mortality_charge_value(238679, 5946, 0.000938, 125)
```

```python
def admin_charge_value(CV_BoY_after_premium_val, admin_charges, admin_inflation, policy_year_val):
    if(CV_BoY_after_premium_val <= 0):
        return 0
    else:
        val = admin_charges * (1 + (admin_inflation/100)) ** (policy_year_val - 1) / 12
        return val
```

```python
def cash_value_EoY(CV_BoY_after_premium_val, mortality_charge_value, admin_charge_val):
    try:
        value = round(CV_BoY_after_premium_val - mortality_charge_value - admin_charge_val)
    except Exception as e:
        value = 0
    return value
```

```python
def investment_income(growth_rate_val, cash_value_eoy_val):
    value = round(cash_value_eoy_val * (growth_rate_val/100))
    return value
```

```python
def policy_holder_account(basic_cash_value, policy_month, total_policy_month, policy_year, sum_assured, death_benefit, mortality_rate, mortality_charge_loading, admin_charges, admin_inflation, growth_rate_val):
    iteration = total_policy_month - policy_month[0]
    cash_val_boY = [round(max(basic_cash_value, 0))]
    base_premium_list = [base_premium[0]]
    allocated_percentage = [allocated_percentage_list[0]]
    allocated_premium_list = [allocated_premium[0]] 
    unallocated_premium_list = [unallocated_premium[0]]
    bid_offer_charge_list = [bid_offer_charge[0]]
    CV_BoY_after_premium = [(max(round(allocated_premium[0] + cash_val_boY[0] - bid_offer_charge_list[0]), 0))]
    sum_at_risk_list = [sum_at_risk_value(death_benefit, sum_assured, CV_BoY_after_premium[0], policy_month[0], total_policy_month)]
    mortality_charge_list = [mortality_charge_value(CV_BoY_after_premium[0], sum_at_risk_list[0], mortality_rate[0], mortality_charge_loading)]
    admin_charge_list = [admin_charge_value(CV_BoY_after_premium[0], admin_charges, admin_inflation, policy_year[0])]
    cash_value_EoY_list = [cash_value_EoY(CV_BoY_after_premium[0], mortality_charge_list[0], admin_charge_list[0])]
    investment_income_list = [investment_income(growth_rate_val, cash_value_EoY_list[0])]
#     table_data = {
#         'Cash Value at BoY': cash_val_boY[0], 
#         'Base Premium': base_premium_list[0], 
#         'Allocated Percentage': allocated_percentage[0], 
#         'Allocated Premium': allocated_premium_list[0],
#         'UnAllocated Premium': unallocated_premium_list[0],
#         'Bid/Offer Charge': bid_offer_charge_list[0],
#         'CV BoY after Premium': CV_BoY_after_premium[0],
#         'Sum at Risk': sum_at_risk_list[0],
#         'Mortality Charge': mortality_charge_list[0],
#         'Admin Charge': admin_charge_list[0],
#         'Cash Value(EoY)': cash_value_EoY_list[0],
#         'Investment Income': investment_income_list[0]
#         }
#     df = pd.DataFrame(table_data)
#     return df
    return cash_val_boY, base_premium_list, allocated_percentage, allocated_premium_list, unallocated_premium_list, bid_offer_charge_list[0], CV_BoY_after_premium, sum_at_risk_list, mortality_charge_list, admin_charge_list, cash_value_EoY_list, investment_income_list
```
